```python
import pandas as pd
```

```python
df = pd.read_csv("exp-1/firewall-pcm.csv", header=[0,1],)
```

```python
df[19970:20010]
```

```python
df["Date_time"] = df[("System", 'Date')] + ' ' + df[("System", 'Time')]
df['Time'] = pd.to_datetime(df["Date_time"])
df['time_difference'] = df['Time'].diff()
df['time_difference_seconds'] = df['time_difference'].dt.total_seconds().astype(float)
```

```python
df['time_difference_seconds'][19900:19982].plot(style=".-")
```
